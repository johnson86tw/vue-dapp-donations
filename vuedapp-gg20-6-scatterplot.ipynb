```python
import json
import pandas as pd
from brokenaxes import brokenaxes
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter


with open('vuedapp-gg20-4.json') as f:
	donations = json.load(f)

df = pd.DataFrame(donations)
df
```

```python
# turn timestamp into datetime
df['datetime'] = pd.to_datetime(df['timestamp'], unit='ms')
df
```

```python
from datetime import datetime
import pytz

plt.figure(figsize=(20, 10), dpi=100)

plt.style.use('ggplot') # for background grid

bax = brokenaxes(ylims=((0, 16), (49, 51), (99, 101)), hspace=.05)

# Create the scatterplot
handles = bax.scatter(
    x=df['datetime'],
    y=df['usd_value_at_time'],
    s=df['usd_value_at_time'] * 10,
    color='#1679AB',
	
)

# x-axis ticks
start_date = df['datetime'].iloc[0]
end_date = df['datetime'].iloc[-1]

date_range = pd.date_range(start=start_date, end=end_date, freq='D')  # Monthly frequency
date_range = date_range.to_pydatetime()  # Convert to datetime objects


bax.axs[0].set_xticks(ticks=date_range) # the upper plot
bax.axs[1].set_xticks(ticks=date_range)
bax.axs[2].set_xticks(ticks=date_range) 

# y-axis ticks
bax.axs[2].set_yticks(range(0, 16, 1)) # the lower plot


bax.set_ylabel("USD at time of donation")
bax.set_title(f"Vue Dapp GG20 Donations (2024/04/23 12:00 - 2024/05/07 23:59 UTC)")



# Legend (not working properly)
# bax.legend(handles=handles, labels=['USD at time'], loc='upper left', title='Donation amount')


date_strings = ["10:46 PM Â· Apr 24, 2024", "5:16 PM Â· Apr 25, 2024", "9:59 PM Â· Apr 28, 2024", "6:18 PM Â· May 3, 2024", "2:37 PM Â· May 6, 2024"]
current_tz = pytz.timezone('Asia/Taipei')  # UTC+8
target_tz = pytz.timezone('America/Chicago')  # CST


for date_string in date_strings:
    date_object = datetime.strptime(date_string, "%I:%M %p Â· %b %d, %Y")
    date_object = current_tz.localize(date_object)
    date_object = date_object.astimezone(target_tz)

    for ax in bax.axs:
        ax.axvline(x=date_object, color='r', linestyle='--')


date_string = "9:20 AM Â· May 3, 2024"
date_object = datetime.strptime(date_string, "%I:%M %p Â· %b %d, %Y")
date_object = current_tz.localize(date_object)
date_object = date_object.astimezone(target_tz)

for ax in bax.axs:
    ax.axvline(x=date_object, color='orange', linestyle='--')
```
