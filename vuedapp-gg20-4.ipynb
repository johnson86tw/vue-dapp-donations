# Get Passport scores

```python
import json
import pandas as pd
from web3 import Web3
from ens import ENS


with open('vuedapp-gg20-3.json') as f:
	donations = json.load(f)

df = pd.DataFrame(donations)
df
```

```python
import requests

records = {}
counter = -1

def get_score(address):
	global counter
	counter += 1
	print(counter)

	url = f"https://api.scorer.gitcoin.co/registry/score/335/{address}"

	if address in records:
		return records[address]

	try:
		response = requests.get(url, headers={
			"Authorization": "Bearer xn9e7AFv.aEfS0ioNhaVtww1jdwnsWtxnrNHspVsS"
		})
		res = response.json()
		score = res['evidence']['rawScore']
		return score
	except:
		return None

score = get_score('0x96A4715280c3Dac3F3093d51aA278aA5eb60ffDE')
score
```

```python
# about 15m 11s

df['score'] = df['donor'].apply(get_score)
df
```

```python
df.to_json('vuedapp-gg20-4.json', orient="records")
```
