# Historical Price

CoinGecko https://www.coingecko.com/zh-tw/%E6%95%B8%E5%AD%97%E8%B2%A8%E5%B9%A3/arbitrum
pycoingecko: https://github.com/man-c/pycoingecko
https://docs.coingecko.com/v3.0.1/reference/coins-id-market-chart

```python
import pycoingecko
import pandas as pd

coinGecko = pycoingecko.CoinGeckoAPI()

eth_data = coinGecko.get_coin_market_chart_by_id('ethereum', 'usd', '30days')
df = pd.DataFrame(eth_data['prices'], columns=['timestamp', 'price'])


df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
df = df[(df['timestamp'] >= '2024-04-23 12:00:00') & (df['timestamp'] <= '2024-05-07 23:59:59')]
df

```

```python
df.to_json('gg20/eth_gg20_historical_price.json', orient='records')
```

```python
import matplotlib.pyplot as plt

# Plot the data
plt.plot(df['timestamp'], df['price'])
plt.xlabel('Date')
plt.xticks(rotation=45)

plt.ylabel('Price (USD)')
plt.title('GG20 Historical ETH Price (USD)')


plt.show()
```
