# Get Donors

```python
import json
import pandas as pd
from web3 import Web3
import os
from dotenv import load_dotenv
load_dotenv()
ALCHEMY_ARBITRUM_API_KEY = os.getenv('ALCHEMY_ARBITRUM_API_KEY')

with open('vuedapp-gg20-1.json') as f:
	donations = json.load(f)

df = pd.DataFrame(donations)
df
```

```python
# get donor
# warning: this is slow. I run 3m 32s for 700 transactions.

web3 = Web3(Web3.HTTPProvider(f'https://arb-mainnet.g.alchemy.com/v2/{ALCHEMY_ARBITRUM_API_KEY}'))

def get_donor(hash):
	tx = web3.eth.get_transaction(hash)
	return web3.to_checksum_address(tx['from'])

df['donor'] = df['hash'].apply(get_donor)
```

```python
df.to_json('vuedapp-gg20-2.json', orient="records")
```
