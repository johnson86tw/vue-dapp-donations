```python
import json
import pandas as pd
from web3 import Web3
from ens import ENS
import os
from dotenv import load_dotenv
load_dotenv()
ALCHEMY_MAINNET_API_KEY = os.getenv('ALCHEMY_MAINNET_API_KEY')


with open('vuedapp-gg20-2.json') as f:
	donations = json.load(f)

df = pd.DataFrame(donations)
df
```

```python
w3 = Web3(Web3.HTTPProvider(f'https://eth-mainnet.g.alchemy.com/v2/{ALCHEMY_MAINNET_API_KEY}'))
ns = ENS.from_web3(w3)

domain = ns.name('0x96A4715280c3Dac3F3093d51aA278aA5eb60ffDE')
domain
```

```python
# warning: this is slow. I run 80m 2s for 700 transactions
ens_names = {}

def get_ens_name(donor):
    # If the name is already in the dictionary, return it
    if donor in ens_names:
        return ens_names[donor]

    # Otherwise, do the lookup and store the result in the dictionary
    try:
        name = ns.name(donor)
        ens_names[donor] = name
        return name
    except Exception:
        ens_names[donor] = None
        return None

df['donor_ens'] = df['donor'].apply(get_ens_name)

df
```

```python
df.to_json('vuedapp-gg20-3.json', orient="records")
```
